```python
import requests
import pandas as pd

URL = ' https://rickandmortyapi.com/api'
```

```python
URL
```

```python
response = requests.get(URL)
response
```

```python
response.status_code
```

```python
data = response.json()
data
```

```python
char_response = requests.get(data['characters'])
char_response
```

```python
# to get characters data
char_data = char_response.json()
char_df = pd.DataFrame(char_data['results'])
char_df.head()
```

```python
#to get locations data
loc_response=requests.get(data['locations'])
loc_data=loc_response.json()
loc_df=pd.DataFrame(loc_data['results'])
```

```python
#to get episodes data
epis_response=requests.get(data['episodes'])
epis_data=epis_response.json()
epis_df=pd.DataFrame(epis_data['results'])
```

**List the names and release years of all the episodes where Summer Smith is present, in chronological order.**

```python

summer_smith_ep=char_df.loc[char_df['name']=='Summer Smith','episode'][2]

for d in summer_smith_ep:
  for j,val in enumerate (epis_df['url']):
    if d==val:
      print(epis_df.loc[j,['name','air_date']])
```

**What are the total episodes in which Jerry Smith appeared?**

```python

jerry_smith_ep=char_df.loc[char_df['name']=='Jerry Smith','episode'][4]

count=0
for d in jerry_smith_ep:
  for j,val in enumerate (epis_df['url']):
    if (d==val):
      count=count+1
      print(epis_df.loc[j,'name'])
print(count)
```

**List all characters created in 2017, in descending order by species. For characters with the same species, order them alphabetically by their name.**

```python

char_df['created']=pd.to_datetime(epis_df['created'])
char_df.loc[char_df['created'].dt.year==2017,['name','species']].sort_values(['species', 'name'],ascending=[False , True ])
```

**List the names of all characters who appeared in an episode in which Amish Cyborg also appeared.**

**List top 5 characters with maximum number of episodes. Start from the highest.**

```python

epi_count_li = []
epis_df['characters']
for i,val in enumerate (epis_df['characters']):
  epi_count_li.append((len(val),i))
epi_count_li.sort(key = lambda i:i[0], reverse = True)

for i,j in epi_count_li[:5]:
  top5_char=epis_df.loc[j,'characters']

l=[]
l.extend(top5_char)

l=set(l)


for i in l:
  for j,val in enumerate (char_df['url']):
    if (i==val):
      print(char_df.loc[j,'name'])

```
